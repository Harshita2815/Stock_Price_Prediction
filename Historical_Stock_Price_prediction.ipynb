```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import numpy as np
# Drop missing values
diff_data = data['LogClose_diff'].dropna()

plt.figure(figsize=(12, 10))

# ACF plot
plt.subplot(1, 2, 1)
plot_acf(diff_data, ax=plt.gca(), lags=40)
plt.title('ACF of Differenced Close Price')

# PACF plot
plt.subplot(1, 2, 2)
plot_pacf(diff_data, ax=plt.gca(), lags=40, method='ywm')
plt.title('PACF of Differenced Close Price')

plt.tight_layout()
plt.show()
```

from here we can conclude that p=0, q=0 and d=1(previously found)

Standard time series practice is to select low values for p and q (typically 0 or 1) when both ACF and PACF show no significant spikes.

ARIMA(0,1,0), ARIMA(1,1,0), and ARIMA(0,1,1) are typically considered in the scenarioâ€”where the ACF and PACF of the differenced series show no significant spikes



ARIMA(0,1,0) means:



1. 0 autoregressive (AR) terms (p=0).

2. 1 order of differencing to stationarize (d=1).

3. 0 moving average (MA) terms (q=0).

4. This model is just a random walk with drift (if a constant is included).

5. We choose this when the differenced series looks like white noise: neither ACF nor PACF has significant spikes, so neither AR nor MA is needed

ARIMA(1,1,0) or ARIMA(0,1,1):

1. If we see a significant spike at lag 1 in PACF (but not in ACF), that points to an AR(1) component, so ARIMA(1,1,0) may be justified.

2. If you see a significant spike at lag 1 in ACF (but not in PACF), that suggests an MA(1) component, so ARIMA(0,1,1) may be better.

```python
from statsmodels.tsa.arima.model import ARIMA
data = data.asfreq('B')  #'B' for business days as weekends are missing

models = {
    'ARIMA(0,1,0)': (0,1,0),
    'ARIMA(1,1,0)': (1,1,0),
    'ARIMA(0,1,1)': (0,1,1),
}

for name, order in models.items():
    model = ARIMA(data['LogClose'], order=order)
    result = model.fit()
    print(f"{name} â†’ AIC: {result.aic:.2f}, BIC: {result.bic:.2f}")
```

The simplest model, ARIMA(0,1,0) (a random walk), has the lowest AIC and BIC.

That makes it the best-fitting model among the three.

The differences in AIC/BIC are small, but consistent.

No AR or MA terms are needed â€” the differenced series behaves like white noise or random walk.

Future value depends only on todayâ€™s value, with no meaningful short-term memory.

```python
import numpy as np
import matplotlib.pyplot as plt

# Forecast the next N steps
n_steps = 30  # example: forecast next 30 business days
forecast_result = result.get_forecast(steps=n_steps)

# Get predicted mean in log-scale
log_forecast = forecast_result.predicted_mean

# Inverse the log transform to get forecast in original scale
forecast = np.exp(log_forecast)

# Get confidence intervals (also in log-scale), and inverse-transform them
conf_int = forecast_result.conf_int()
conf_int_exp = np.exp(conf_int)

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Historical Close')
plt.plot(forecast.index, forecast, label='Forecast', color='green')
plt.fill_between(forecast.index,
                 conf_int_exp.iloc[:, 0],
                 conf_int_exp.iloc[:, 1],
                 color='lightgreen', alpha=0.5, label='95% Confidence Interval')

plt.title('Forecast of Maruti Stock Prices (ARIMA(0,1,0))')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Apply 1D CNN-LSTM

Prepare the dataset

We need to convert your time series into a supervised learning format (with input sequences and targets).

```python
window_size = 30  # you can tune this
```

```python
import numpy as np

def create_sequences(series, window_size):
    X = []
    y = []
    for i in range(len(series) - window_size):
        X.append(series[i:i+window_size])
        y.append(series[i+window_size])
    return np.array(X), np.array(y)

# Use log-transformed series if you want
series = np.log(data['Close'].dropna()).values
X, y = create_sequences(series, window_size)

# Reshape for CNN-LSTM: [samples, timesteps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
train_size = int(0.7 * len(X))
val_size = int(0.15 * len(X))

X_train, y_train = X[:train_size], y[:train_size]
X_val, y_val = X[train_size:train_size + val_size], y[train_size:train_size + val_size]
X_test, y_test = X[train_size + val_size:], y[train_size + val_size:]
```

CNN-LSTM model

```python
from tensorflow.keras import Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense

model = Sequential([
    Input(shape=(window_size, 1)),
    Conv1D(filters=64, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    LSTM(50),
    Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1)
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=32,
    callbacks=[early_stop, reduce_lr]
)
```

```python
# Predict on test set
y_pred = model.predict(X_test).squeeze()

# Reverse log-transform
y_pred_orig = np.exp(y_pred)
y_test_orig = np.exp(y_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test_orig, y_pred_orig)
mae = mean_absolute_error(y_test_orig, y_pred_orig)
r2 = r2_score(y_test_orig, y_pred_orig)
n, p = len(y_test_orig), X_test.shape[1]
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RÂ²: {r2:.4f}")
print(f"Adjusted RÂ²: {adj_r2:.4f}")
```
